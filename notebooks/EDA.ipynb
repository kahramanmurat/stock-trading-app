### Understanding Financial Data

```python
import pandas as pd
```

```python
df = pd.read_csv('sp500full.csv')
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
df.info()
```

```python
df['name'].unique()
```

```python
df['name'].unique().shape
```

```python
df[df['name'] == 'AAPL']
```

```python
aapl = df[df['name'] == 'AAPL']
aapl['close']
```

```python
aapl['close'].plot()
```

```python
# convert to timestamp object
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date'].min(), df['date'].max()
```

```python
dates = pd.date_range(df['date'].min(), df['date'].max())
dates
```

```python
close_prices = pd.DataFrame(index=dates)
```

```python
close_prices.head()
```

```python
symbols = df['name'].unique()
```

```python
df2 = pd.DataFrame(data=aapl['close'].to_numpy(), index=aapl['date'], \
                   columns=['AAPL'])
```

```python
df2.head()
```

```python
# Let's now fill in our close prices dataframe
for symbol in symbols:
  df_sym = df[df['name'] == symbol]
  df_tmp = pd.DataFrame(data=df_sym['close'].to_numpy(), \
                        index=df_sym['date'], columns=[symbol])
  close_prices = close_prices.join(df_tmp) # left-join by default
```

```python
close_prices.head()
```

```python
close_prices.info()
```

```python
close_prices.to_csv('sp500_close.csv')
```

```python
# Now it has dates
# Also has missing values
close_prices['AAPL'].plot()
```

```python
# drop rows with all nan
# most likely correspond to weekends, holidays (non-trading days)
close_prices.dropna(axis=0, how='all', inplace=True)
```

```python
close_prices['AAPL'].plot()
```

```python
close_prices.isna().sum()
```

```python
close_prices.isna().sum().sum()
```

```python
close_prices.iloc[0, :].isna().sum()
```

```python
close_prices.fillna(method='ffill', inplace=True)
```

```python
close_prices.isna().sum().sum()
```

```python
close_prices.fillna(method='bfill', inplace=True)
```

```python
close_prices.isna().sum().sum()
```

```python
close_prices.plot(legend=False, figsize=(10, 10));
```

```python
close_prices_normalized = close_prices / close_prices.iloc[0]
```

```python
close_prices_normalized.plot(legend=False, figsize=(10, 10));
```
